Ejemplos While

```
n = 10
while (n > 0):
  print(n)
  n = n - 1
print("Despegue...")
```

```
while True:
  linea = input('>')
  if linea == 'exit':
    break
  print(linea)
print('Terminado')
```

```
while True:
  linea = input('>')
  if linea[0] == '#':
    continue
  if linea == 'exit':
    break
  print(linea)
print('Terminado')
```

Ejemplos for


```
alumnos = ['Carlos', 'Guillermo', 'Luis', 'Manuel', 'Martin', 'Ricardo']
for alumno in alumnos:
  print('Hola te invito al curso de programacion, Saludos:', alumno)
print('Terminado')
```

```
for i in range(10):
  print(i)
print('Terminado...')
```

```
for i in range(1,10):
  print(i)
print('Terminado...')
```

```
# Generar una lista del 0 al 200 de 10 en 10
for i in range(0,201,10):
  print(i)
print('Terminado..')
```

```
# Sumador de elementos dentro de una coleccion
vector =[52,63,14,20,32,56,74,15,14,23]
suma = 0
for valor in vector:
  suma = suma + valor
  print(suma)
print('Total =', suma)
```

```
sum(vector)
```

```
# Mayor elemento dentro de una coleccion
mayor = None
for valor in vector:
  if mayor is None or valor > mayor:
   mayor = valor
  print('Bucle',valor,mayor)
print('Valor Maximo; ', mayor)
```

```
max(vector)
```

```
# Menor elemento dentro de una coleccion
menor = None
for valor in vector:
  if menor is None or valor < menor:
    menor = valor
  print('Bucle',valor,menor)
print('Valor minimo: ', menor)
```

```
min(vector)
```

```
# Calcular la media de una lista de numeros ingresada por el usuario

sum = 0;
cont = 0;
while True:
  num = float(input('>'))
  if num == ( 0 ):
    break
  sum = sum + num
  cont = cont + 1
print('Suma total =', sum)
print('Elementos Ingresados = ', cont)
print('Media = ', sum/cont)
```
