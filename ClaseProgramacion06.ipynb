```
Alumnos4 = {'Juan','Pedro','Sheyla','Manuel','Luis','Daniel','Armando','Daniela','Carlos'}
Ingles = {'Juan','Pedro','Daniela','Manuel','Rodrigo','José','Pamela'}
Musica = {'Carlos','Daniel','Daniela','Pedro','Luis','José','Natalia','Marta'}
```

```
# Alumnos de 4to que están en el taller de Música
Alumnos4 & Musica
```

```
# Alumnos que Cursan Inglés y Musica
Ingles & Musica
```

```
# Alumnos que Cursan Musica y no van en 4to
Musica - Alumnos4
```

```
#Alumnos de 4to que cursan musica e Ingles
Alumnos4 & Ingles & Musica
```
