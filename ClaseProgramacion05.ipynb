# Ejercicios

```
# Ejercicio 1
print("Programa que Calcula Área y Perímetro de una Circunferencia dado su radio")
r = float(input("\nIngrese el radio de la circunferencia = "))

A = math.pi*r**2 # pi*r^2
P = 2*math.pi*r
A = round(A,2); P = round(P,2) 

print('El Área de la Círculo es: ', A)
print('El Perímetro del Círculo es: ', P)
```

```
# Ejercicio 2
print('Programa que Calcula el número de diagonales y la suma de angulos internos de un Poligono de n lados')
n = int(input('\nIngrese el número de lados = '))

D = n*(n-3) / 2
S = 180 * (n-2)

print('El número de Diagonales trazables en el Poligono es: ', D)
print('La Suma de los Ángulos Internos del Poligono es: ', S)
```
