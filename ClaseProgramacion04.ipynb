  - Consulta de saldo
  - Depositos
  - Retiros
  - Cambio de NIP
  - Pago de servicios

```
print('Bienvenido al Cajero ATM de Banco de Mexico')
print('-------------------------------------------')
nip = '2054'
NoCu = '123456789'
saldo = 35000.00
celA = '4621234567'

NIP = input('\nIngrese su NIP: ')
while (nip == NIP):
  print('\n*****Menu de Opciones*****')
  print('\n1. Consulta de Saldo')
  print('2. Realizar un deposito')
  print('3. Realizar un Retiro')
  print('4. Cambio de NIP')
  print('5. Operacion Finalizada')
  print('**************************')

  pick = int(input('\nIngrese la opcion deseada: '))

  if (pick == 1):
    print('\nConsulta de Saldo--------------')
    print('\nNumero de Cuenta: ', NoCu)
    print('\nSaldo actual: $', saldo)
  elif (pick == 2):
    print('\nDepositos a tu cuenta----------')
    dep = float(input('Cantidad a depositar: $'))
    saldo = saldo + dep
    print('\nTu saldo actual es: $ ', saldo)
  elif (pick == 3):
    print('\nRetiros------------------------')
    ret = int(input('Cantidad a retirar: $'))
    saldo = saldo - ret
    print('Favor de retirar tu dinero')
    print('\n Tu saldo actual es: $', saldo)
  elif (pick == 4):
    print('\nCambio de NIP------------------')
    cel = input('\nIngresa tu numero Celular: ')
    if (celA == cel):
      print('\nHemos enviado un mensaje de texto a tu celular')
      clave = input('\nIngrese la clave que hemos enviado: ')

      NIP = input('Ingrese su nuevo NIP (4 Digitos): ')
      print('\nNIP Actualizado')
    else:
      print('\nEl numero de celular no coincide, intentalo en otra ocasion')
  elif (pick == 5):
    break
  else:
    print('\nIngrese por favor la opcion correcta')

print('\nGracias por Visitarnos. No olvides retirar tu tarjeta')
```
