Ejercicio 1. Convertidor de Unidades en Temperatura

```
print('Programa convertidor de temperatura (C--->F)')
print('===========================================')
C = float(input('Ingrese la temperatura en Grados Celsius = '))

F = 9*C/5 + 32

print("La temperatura en F equivale a: ", F)
```

```
#Ingreso a un parque de Diversiones a Mayores de Edad

edad = int(input('Ingresa su edad: '))

if ( edad >= 18 ): 
  print('Bienvenid@. ¡Diviertete!')
else:
  print('No cumples con la mayoria de edad')

print('Hasta la proxima')
```

if....elif....else

```
#Programa que te asigna a tu generacion
year = int(input('Ingrese su año de nacimiento: '))

if ( year >= 1946 and year <= 1964 ):
  print('Año de nacimiento: ', year)
  print("Usted pertenece a la generacion Baby Boomers")
elif ( year >= 1965 and year <= 1980 ):
  print('Año de nacimiento: ', year)
  print("USted pertenece a la generacion X")
elif ( year >= 1981 and year <= 2000 ):
  print('Año de nacimiento: ',year)
  print("Usted pertence a la generacion Y: Milenial")
elif ( year >= 2001 and year < 2021 ):
  print('Año de nacimiento: ', year)
  print("Usted pertenece a la generacion Z: Boomlets")
else:
  print('Año de nacimiento: ',year)
  print("El año que ingresaste esta fuera de las generaciones")

print('Hasta la proxima....')
```

```
# Detector de Pares e Impares

print("Programa que distingue pares de impares")
print("=======================================")
num = float(input('Ingrese el numero: '))
num = round(num)
a = num % 2
if ( a == 0 or a != 0 ):
  print('El numero ', num)
  print('Es par')
else:
  print('El numero ', num)
  print('Es impar')
```
